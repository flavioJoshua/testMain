```python
import numpy as np
from scipy.stats import gaussian_kde

# Dati delle serie
serie1 = np.array([63, 6, 23, 32, 45, 48])
serie2 = np.array([1.45, -1.63, -0.71, -0.23, 0.48, 0.64])

# Calcolo della stima della densit√† del kernel per entrambe le serie
kde1 = gaussian_kde(serie1)
kde2 = gaussian_kde(serie2)

# Intervallo dei valori per il quale calcolare la PDF
x1 = np.linspace(min(serie1) - 10, max(serie1) + 10, 100)
x2 = np.linspace(min(serie2) - 1, max(serie2) + 1, 100)

# Calcolo della PDF
pdf1 = kde1(x1)
pdf2 = kde2(x2)

print("Valori PDF Serie 1:", pdf1)
print("Valori PDF Serie 2:", pdf2)
```
